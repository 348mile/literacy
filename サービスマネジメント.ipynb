```python
import pandas as pd
import folium

df = pd.read_excel("12_sports.xls")
display(df)

sports = df[["緯度","経度","施設名称"]].values
print(sports)

m = folium.Map(location=[35.66206, 140.0141],zoom_start=15)
for data in sports:
  folium.Marker([data[0],data[1]],tooltip=data[2]).add_to(m)
folium.Circle(location=[35.66206, 140.0141], radius=1_000).add_to(m)
display(m)
```
